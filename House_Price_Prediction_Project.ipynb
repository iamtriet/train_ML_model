### Feature selection

```python
# Create X (After completing the exercise, you can return to modify this line!)
features = [
    'LotArea',
    'YearBuilt',
    '1stFlrSF',
    '2ndFlrSF' ,
    'FullBath',
    'BedroomAbvGr',
    'TotRmsAbvGrd'
]
```

### 3. Splitting dataset into X and y

```python
X = data[features]
y = data["SalePrice"]
```

### X, y &#8594; X_train, y_train, X_valid, y_valid

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

### 4. Training Machine Learning Model

```python
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor(random_state=1)
```

```python
#Fit training data into model
dt_model.fit(X_train, y_train)
```

```python
y_preds = dt_model.predict(X_valid.head())
```

```python
y_preds
```

```python
pd.DataFrame({'y': y_valid.head(), 'y_preds': y_preds})
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(X_train, y_train)
```

```python
rd_val_preds = rf_model.predict(X_valid)
```
